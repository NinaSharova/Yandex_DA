## Предобработка данных

```python
print(f'Всего пропусков до:\n{data.isna().sum()}') # зафиксируем количество пропусков до обработки
```

```python
print(f'Всего пропусков до:\n{data.isna().mean()}') # зафиксируем долю пропусков до обработки
```

* Третью, по величине, долю пропусков составляет количество балконов: вероятно для объектов, у которых отсутствуют балконы, не указано значение 0. Заполним такие пропуски на 0.

* Одинаковую долю пропусков составляют данные о расстоянии до близлежащих аэропортов, до центра города, наличие парков и водоемов в радиусе 3 км: необходимо проверить, а какие данные вообще внесены для этих строк.

```python
# проверим столбцы на артефакты
display(data.describe())
```

Отрицательные значения не найдены, однако строки с максимальной площадью и стоимостью, и наоборот - с высостой потолков в 1 метр и стоимостью 12 тысяч рублей, вызывают вопросы о корректности внесенных данных.

```python
# заменим пропуски в is_apartment на False
data['is_apartment'] = data['is_apartment'].fillna(False)
```

```python
# заменим пропуски в balcony на 0
data['balcony'] = data['balcony'].fillna(0)
```

```python
# заполним вспомогательный столбец информацией о факте пропуска
data['no_date_days_exposition'] = data['days_exposition'].isna() 

# составим срез данных для строк, у которых не заполнены данные о количестве дней размещения объявления
display(data.query('no_date_days_exposition == True').head(10))
```

В результате построения среза не удалось установить прямую зависимость между датой размещения объявления и отсуствием данных в столбце о количестве дней с момента публикации.

Однако фильтрация показа другую интересную закономерность - для объявлений, в которых не указана длительность публикации, зачастую отсуствуют значения в других столбцах, важных для исследования (ceiling_height, living_area, cityCenters_nearest).

```python
# заполним вспомогательные столбцы фактами о наличии пропуска
data['no_ceiling_height'] = data['ceiling_height'].isna() 
data['no_living_area'] = data['living_area'].isna()
data['no_cityCenters_nearest'] = data['cityCenters_nearest'].isna()

# сделаем срез и посчитаем количество строк с пустыми значениями в трех значимых столбцах
print('Количество строк с пропусками в значимых столбцах:', 
      data
        .query('no_ceiling_height == True and no_living_area == True and no_cityCenters_nearest == True')
        .shape[0]
     ) 
```

Имеются строки сразу с несколькими пропусками в значимых для исследования столбцах. Однако количество этих строк позволяет пренебречь ими в рамках исследования. Требуется передать запрос техническим специалистам, осуществлявшим выгрузку, о возможных причинах появления целых незаполненных строк, а не отдельных ячеек.

```python
print(f'Всего пропусков после:\n{data.isna().sum()}') # зафиксируем количество пропусков после обработки
```

## Расчёты и добавление результатов в таблицу

Проведем рассчёты и добавим с таблицу: цену квадратного метра, день недели, год и месяц публикации объявления, этаж (первый, последний, другой), соотношение площадей (жилой к общей, кухни к общей).

```python
#рассчитаем и запишем в новый столбец цену квадратного метра
data['price_per_m'] = data['last_price'] / data['total_area']
```

```python
#проведем рассчеты для даты и времени
data['weekday_exposition'] = data['first_day_exposition'].dt.weekday
data['month_exposition'] = data['first_day_exposition'].dt.month
data['year_exposition'] = data['first_day_exposition'].dt.year
```

```python
#Заменим числовые значения в днях недели
data = data.replace({'weekday_exposition':
                     {0:'пн', 
                      1:'вт',
                      2:'ср',
                      3:'чт',
                      4:'пт',
                      5:'сб',
                      6:'вс'}})
```

```python
# категоризириуем этаж

#напишем функцию, которая принимает два аргумента - этаж квартиры и всего этажей в доме
#если этаж равен первому, функция вернет "первый"
#если этаж равен общему количеству этажей, функция вернет "последний"
#в остальных случая функция вернет "другой"

def what_floor(row_values):
    if row_values['floor'] == 1:
        return 'первый'
    if row_values['floor'] == row_values['floors_total']:
        return 'последний' 
    return 'другой'

data['category_floor'] = data.apply(what_floor, axis=1) # применим функцию к ряду датафрейма целиком
```

```python
#посчитаем соотношение площадей
data['total_to_living_area'] = data['living_area'] / data['total_area'] 
data['kitchen_to_total_area'] = data['kitchen_area'] / data['total_area'] 
```

- month_exposition - месяц публикации объявления
- year_exposition - год публикации объявления
- category_floor - категория этажа
- total_to_living_area - соотношение жилой и общей площади
- kitchen_to_total_area - соотношение площади кухни и общей площади

```python
#напишем функцию, которая будет строить график вида "boxplot" для набора столбцов
def boxplot_to_data(data_to_plot, column_to_plot, name, x_name, y_name):
    data_to_plot.boxplot(column=column_to_plot, grid=True, figsize =(4, 7))
    plt.title(name)
    plt.xlabel(x_name)
    plt.ylabel(y_name)
    plt.show()
```

```python
#напишем функцию, которая будет строить график вида "boxplot" для набора столбцов
def hist_to_data(data_to_plot, column_to_plot, name, x_name, y_name):
    data_to_plot.hist(column=column_to_plot, grid=True, bins=30)
    plt.title(name)
    plt.xlabel(x_name)
    plt.ylabel(y_name)    
    plt.show()
```

### Исследование факторов, влияющих на стоимость квартиры <a name='4.6._факторы'></a>

Сформулируем гипотезу: существует зависимость между площадью квартиры, числом комнат, этажом, удаленностью от центра и её ценой. Проверим это предположение. Для этого:

* посчитаем коэфициент корреляции для каждой пары параметров (цена-площадь, цена-комнаты, цена-этаж, цена - удаленность),
* построим диаграммы рассения графики, поделённый на шестиугольные области для визуализации зависимости (на основе "очищенной таблицы");
* при необходимости построим графики на основе сводных таблиц.

```python
#напишем функцию, которая будет выводить коэффициент корреляции для набора столбцов
def corr_column_to_column(column_1, column_2):
    result = column_1.corr(column_2)
    return print('По таблице, построенной без учета выбросов и аномалий, коэффициент корреляции составляет {:.2}'
                 .format(result))
```

```python
#напишем функцию, которая будет строить график вида "hexbin" для набора столбцов
def hexbin_to_columns(data_to_plot, name, x_name, y_name):
    data_to_plot.plot(data_to_plot.columns.values[0], data_to_plot.columns.values[1], kind='hexbin', gridsize=20, figsize=(8, 6), sharex=False, grid=True)
    plt.title(name)
    plt.xlabel(x_name)
    plt.ylabel(y_name)
    plt.show()
```

```python
#напишем функцию, которая будет строить график зависимости для набора столбцов
def scatter_to_columns(data_to_plot, name, x_name, y_name):
    data_to_plot.plot(data_to_plot.columns.values[0], data_to_plot.columns.values[1], kind='scatter', alpha=0.03, grid=True)
    plt.title(name)
    plt.xlabel(x_name)
    plt.ylabel(y_name)
    plt.show()
    
```

```python
#построим графики
hexbin_to_columns(no_strange_data[['total_area','last_price_millions']], 'Зависимость между общей площадью квартиры и ценой', 'Общая площадь (кв.м.)', 'Цена (млн.р.)')

scatter_to_columns(no_strange_data[['total_area','last_price_millions']], 'Зависимость между общей площадью квартиры и ценой', 'Общая площадь (кв.м.)', 'Цена (млн.р.)')
plt.show()
```

```python
#при помощи query() получим срез данных по центру Санкт-Петербурга
center_spb = flats_spb.query('city_center_km <= 4')
```

```python
#напишем функцию, которая будет строить диаграммы для столбца
def hist_to_center_spb(data_to_plot, column_to_plot, name, x_name, y_name):
    data_to_plot.hist(column=column_to_plot, grid=True, bins=30)
    plt.title(name)
    plt.xlabel(x_name)
    plt.ylabel(y_name)    
    plt.show()
```

```python
hist_to_center_spb(center_spb, ['total_area'], 'Гистограмма по столбцу "Общая площадь"', 'Общая площадь (кв.м.)', 'Количество объявлений')
hist_to_center_spb(center_spb, ['rooms'], 'Гистограмма по столбцу "Количество комнат"', 'Количество комнат (шт.)', 'Количество объявлений')
hist_to_center_spb(center_spb, ['last_price_millions'], 'Гистограмма по столбцу "Цена"', 'Цена (млн.р.)', 'Количество объявлений')
hist_to_center_spb(center_spb, ['ceiling_height'], 'Гистограмма по столбцу "Высота потолка"', 'Высота потолка (м.)', 'Количество объявлений')
```

```python
#напишем функцию, которая будет строить график вида "scatter" для набора столбцов
def scatter_to_center(data_to_plot, name, x_name, y_name):
    data_to_plot.plot(data_to_plot.columns.values[0], data_to_plot.columns.values[1], kind='scatter', alpha=0.3, grid=True)
    plt.title(name)
    plt.xlabel(x_name)
    plt.ylabel(y_name)
    plt.show()
```

```python
#напишем функцию, которая будет выводить коэффициент корреляции для набора столбцов
def center_corr_column_to_column(column_1, column_2):
    result = column_1.corr(column_2)
    return print('Коэффициент корреляции составляет {:.2}'
                 .format(result))
```

Проанализируем зависимость межу ценой, количеством комнат, этажом, расстоянием до центра города.

Построим диаграмму рассеяния, сводную таблицу и посчитаем корреляцию между количестом комнат и ценой.

```python
#применим функцию корреляции для пары цена-количество комнат
center_corr_column_to_column(center_spb['rooms'], center_spb['last_price'])

#построим диаграмму рассеяния для пары значений цена-количество комнат
scatter_to_center(center_spb[['rooms','last_price_millions']], 'Диаграмма рассения для параметров:\n зависимость между ценой и количеством комнат','Количество комнат','Цена (млн.р.)')

#построим сводную таблицу
(
    center_spb.pivot_table(index='rooms', values='last_price_millions')
              .plot(grid=True, figsize=(8,6))
)
plt.title('График изменения цены в зависимости от количества комнат в квартире')
plt.xlabel('Количество комнат')
plt.ylabel('Цена (млн.р.)')
plt.show()
```

Коэффицент корреляции указывает на слабую положительную зависимость между ценой и количеством комнат. На диаграмме рассения виден частотный разброс цен между квартирами с разным количеством комнат. Двухкомнатных квартир продается в центре больше всего. По графику сводной таблицы видно увеличение увеличение средней цены в зависимости от количества комнат.

```python
#применим функцию корреляции для пары цена-этаж
center_corr_column_to_column(center_spb['floor'], center_spb['last_price'])

#построим диаграмму рассеяния для пары значений цена-этаж
scatter_to_center(center_spb[['floor','last_price_millions']], 'Диаграмма рассения для параметров:\n зависимость между ценой и этажом','Этаж','Цена (млн.р.)')

#построим сводную таблицу
(
    center_spb.pivot_table(index='category_floor', values='last_price_millions')
              .plot(grid=True, figsize=(8,6))
)
plt.title('График изменения цены в зависимости от этажа')
plt.xlabel('Этаж')
plt.ylabel('Цена (млн.р.)')
plt.show()   
```

Коэффициент корреляции стремиться к нулю, что характеризует осуствие зависимости. По диаграмме рассеяния видно частное распределение квартир на разных этажах. Меньше всего квартир продается на первом этаже. Судя по данным сводной таблицы, квартиры на первом этаже также дешевле, при этом квартиры на последнем и других этажах практически одинаковы в цене.

Проанализуем зависимость между ценой и расстоянием до центра.

```python
#применим функцию корреляции для пары цена-расстояние до центра
center_corr_column_to_column(center_spb['cityCenters_nearest'], center_spb['last_price'])

#построим диаграмму рассеяния для пары значений цена-расстояние до центра
scatter_to_center(center_spb[['city_center_km','last_price_millions']], 'Диаграмма рассения для параметров:\n зависимость между ценой и расстоянием до центра города','Расстояние до центра (км)','Цена (млн.р.)')
plt.show()

#построим сводную таблицу
(
    center_spb.pivot_table(index='city_center_km', values='last_price_millions')
              .sort_values(by='city_center_km')
              .plot(grid=True, figsize=(8,6))    
)
plt.title('График изменения цены в зависимости от удаленности от центра города')
plt.xlabel('Удаленность от центра города (км)')
plt.ylabel('Цена (млн.р.)')
plt.show()
```

По сводной таблице видно незначительное изменение цены, которое непропорционально удалению от центральной точки. В выборке больше всего продающихся квартир на расстоянии 4 км от центра.
